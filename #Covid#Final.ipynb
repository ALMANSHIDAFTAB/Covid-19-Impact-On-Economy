#  Aggregating the data (294) is the mode !

```python
code = data["CODE"].unique().tolist()
country = data["COUNTRY"].unique().tolist()
hdi = []
tc = []
td = []
sti = []
population = data["POP"].unique().tolist()
gdp = []

for i in country:
    hdi.append((data.loc[data["COUNTRY"] == i, "HDI"]).sum()/294)
    tc.append((data2.loc[data2["location"] == i, "total_cases"]).sum())
    td.append((data2.loc[data2["location"] == i, "total_deaths"]).sum())
    sti.append((data.loc[data["COUNTRY"] == i, "STI"]).sum()/294)
    population.append((data2.loc[data2["location"] == i, "population"]).sum()/294)

aggregated_data = pd.DataFrame(list(zip(code, country, hdi, tc, td, sti, population)), 
                               columns = ["Country Code", "Country", "HDI", 
                                          "Total Cases", "Total Deaths", 
                                          "Stringency Index", "Population"])
print(aggregated_data.head())
```

# ML Model To Predict No. Of Deaths With Respect To Population And Total Cases

# Taking ('Total Cases','Total Deaths','Population') From (data) dataset Into data_new Variable

```python
data_new=data[['Total Cases','Total Deaths','Population']]
```

# Resetting  Index!

```python
data_new.reset_index(drop=True)
```

# Importing Necessary Libraries From Training And Testing The Data 

```python
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
```

# Storing Data of data_new In 'X' variable Dropping 'Total Deaths' .


# Storing 'Total Deaths' in 'Y' Variable.

```python
X = data_new.drop(['Total Deaths'],axis=1)
Y = data_new['Total Deaths']
```

# Splitting The data Into 3:7 ratio, 30% testing rest training. 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_test.shape
```

```python
X_train.shape
```

# Scaling The Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
import math
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Importing Linear Regression Model

```python
model = linear_model.LinearRegression()
```

# Fitting The Data Into Model

```python
 model.fit(X_train,Y_train)
```

# Accuracy Check!

```python
model.score(X_train,Y_train)
```

```python
predict=model.predict(X_test)
```

```python
mae=mean_absolute_error(Y_test,predict)
```

```python
mae
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```
